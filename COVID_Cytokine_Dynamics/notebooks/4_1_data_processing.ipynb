```python
import pandas as pd
xls = pd.ExcelFile('../data_raw/adj9285_Data_file_S1.xlsx')
```

```python
import numpy as np
def excel_to_df (df1,tests_indices,patients_cols,patients_infected,len_list,diff_col = 2):
    all_data_df = pd.DataFrame(columns = ['patient.number','infected.yn','type.sample','test','value','dpi'])
    for test_index in tests_indices:
        test_indices = np.arange(test_index,test_index+len_list)
        if test_index+len_list>len(df1):
            test_indices = np.arange(test_index,len(df1))
        for col_num in patients_cols:
            for ind in test_indices:
                if col_num<=patients_infected:
                    infected_yn = 'y'
                else:
                    infected_yn='n'
                row_dict = {'patient.number':[col_num-diff_col],
                            'infected.yn':[infected_yn],
                            'type.sample':df1.loc[0][0],
                            'test':[df1.loc[test_index][0]],
                            'value':[df1.loc[ind][col_num]],
                            'dpi':[df1.loc[ind][0]]
                            }
                all_data_df = pd.concat([all_data_df,pd.DataFrame(row_dict)],ignore_index=True)
    all_data_df = all_data_df.dropna()
    return all_data_df
```

```python
df1 = pd.read_excel(xls, 'Fig2a',header=None)
tests_indices = [1,21,41,61,81,101,121,141]
len_list = 19
patient_cols = np.arange(3,36)
all_data_df = excel_to_df(df1,tests_indices,patient_cols,19,len_list)
print(len(all_data_df))
print(all_data_df['dpi'].unique())
```

```python
def viral_titer_to_df (df1,test_type_indices,num_patients):
    df_ret =  pd.DataFrame(columns = ['patient.number','infected.yn','type.sample','test','value','dpi'])
    for test_index in test_type_indices:
        patient_num = 0
        for index in np.arange(test_index+2,min(test_index+2+num_patients+1,len(df1))):
            patient_num += 1
            for col_num in range(1,38):
                if pd.isna(df1.loc[index,col_num]) == False:
                    try:
                        row_to_add = {'patient.number':[patient_num],
                                    'infected.yn' : ['y'],
                                    'type.sample' : [df1.loc[test_index][0]],
                                    'test': [df1.loc[0][0]],
                                    'value':[df1.loc[index][col_num]],
                                    'dpi':[df1.loc[test_index+1][col_num]]
                                    }
                        df_ret = pd.concat([df_ret,pd.DataFrame(row_to_add)],ignore_index=True)
                    except:
                        continue
    return df_ret
# print(len(pd.read_excel(xls, 'Fig1a',header=None)))
df_ret_5 = viral_titer_to_df(pd.read_excel(xls, 'Fig1a',header=None),[1,22],18)
all_data_df = pd.concat([all_data_df,df_ret_5])
print(all_data_df['dpi'].unique())
print(len(all_data_df))
```

```python
import numpy as np
def excel_to_df2 (df1,tests_indices,patients_cols,patients_infected,len_list,sample_type,
                  diff_cols = 2, row_diff = 0):
    all_data_df = pd.DataFrame(columns = ['patient.number','infected.yn','type.sample','test','value','dpi'])
    for test_index in tests_indices:
        test_indices = np.arange(test_index+row_diff,test_index+row_diff+len_list)
        if test_index+row_diff+len_list>len(df1):
            test_indices = np.arange(test_index+row_diff,len(df1))
        for col_num in patients_cols:
            for ind in test_indices:
                if col_num<=patients_infected:
                    infected_yn = 'y'
                else:
                    infected_yn='n'
                row_dict = {'patient.number':[col_num-diff_cols],
                            'infected.yn':[infected_yn],
                            'type.sample':sample_type,
                            'test':[df1.loc[test_index][0]],
                            'value':[df1.loc[ind][col_num]],
                            'dpi':[df1.loc[ind][0]]
                            }
                all_data_df = pd.concat([all_data_df,pd.DataFrame(row_dict)],ignore_index=True)
    all_data_df = all_data_df.dropna()
    return all_data_df

df7 = pd.read_excel(xls, 'Fig S1a',header=None)
df_ret_7 = excel_to_df2(df7,[0,19,38,57,76,95,114,133,152,171,209,228,247,266,285,304,323,342,361,380,399,418,437,456,475,494],
                        patients_cols=np.arange(1,35),patients_infected=19,len_list=18,sample_type='NASAL',
                        diff_cols = 0,row_diff = 1)
all_data_df = pd.concat([all_data_df,df_ret_7],ignore_index=True)
print(len(all_data_df))
print(all_data_df['dpi'].unique())
```
